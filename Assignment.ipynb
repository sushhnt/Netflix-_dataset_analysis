# Importing Libraries

```python
# Import all the libraries which will be used later
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns       
import numpy as np
%matplotlib inline 
```

# Importing the dataset

```python
# load in data 
df=pd.read_csv('titles.csv')
df
```

# Data Wrangling

```python
df.isnull().sum() #Checking Null Values 
```

```python
df['description'].fillna('Description not available',inplace=True)
df['age_certification'].fillna('Certification not available',inplace=True)
df['seasons'].fillna('Movies do not have seasons',inplace=True) #Replacing Null Values With Appropriate Values
```

```python
df
```

```python
df.isnull().sum() #Checking null values again
```

```python
del df['imdb_id'] #Delete Unuseful Columns 
```

```python
del df['tmdb_score'] #Delete Unuseful Columns 
```

```python
df['imdb_score'].fillna(0,inplace=True) # Filling 0 for null values 
```

```python
df['imdb_votes'].fillna(0,inplace=True) # Filling 0 for null values 
```

```python
df['tmdb_popularity'].fillna(0,inplace=True)# Filling 0 for null values 
```

```python
df['title'].fillna('Not Available',inplace=True)
```

```python
df.isnull().sum() #No null values present now.
```

```python
df
```

```python
df['production_countries'].replace('[]','No data',inplace=True) # Working on countries column
```

```python
df['production_countries'].replace("['US']",'US',inplace=True) # Working on countries column
```

```python
df['production_countries'].replace("['GB']",'GB',inplace=True) # Working on countries column
```

```python
df['production_countries'].replace("['JP']",'JP',inplace=True)
df['production_countries'].replace("['ES']",'ES',inplace=True)
df['production_countries'].replace("['FR']",'FR',inplace=True)
df['production_countries'].replace("['KR']",'KR',inplace=True)
df['production_countries'].replace("['IN']",'IN',inplace=True)
df['production_countries'].replace("['NG']",'NG',inplace=True)
df['production_countries'].replace("['CA']",'CA',inplace=True)
df['production_countries'].replace("['MX']",'MX',inplace=True)
df['production_countries'].replace("['BR']",'BR',inplace=True)
df['production_countries'].replace("['PH']",'PH',inplace=True) # Working on countries column
```

```python
df
```

# Number of movies released each year

```python
movie_count = df['release_year'].value_counts()
# movie_count.plot(xlabel='year',ylabel='number of movies',title='Number of Movies Released Each Year')
fig = plt.figure()
plt.plot(movie_count)
fig.suptitle('Number of Movies Released Each Year',fontsize=20)
plt.xlabel('year',fontsize=18)
plt.ylabel('number of movies',fontsize=18)
```

Number of movies released each year are increasing rapidly especially between 2010-2020

# Percentage of Movies and Shows

```python
plt.figure(figsize=(12,6))
plt.title("Percentation of Netflix Titles that are either Movies or TV Shows")
g = plt.pie(df.type.value_counts(),labels=df.type.value_counts().index, colors=['red','yellow'],autopct='%1.1f%%')
plt.show()
```

# Top 15 countries contributor on Netflix

```python
filtered_countries = df.set_index('title').production_countries.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
filtered_countries = filtered_countries[filtered_countries != 'Country Unavailable']
plt.figure(figsize=(13,7))
g = sns.countplot(y = filtered_countries, order=filtered_countries.value_counts().index[:15])
plt.title('Top 15 Countries Contributor on Netflix')
plt.xlabel('Titles')
plt.ylabel('Production Countries')
plt.show()
```

US produces the highest number of movies and shows followed by IN and JP

# Top 10 Genres on Netflix

```python
filtered_genres = df.set_index('title').genres.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
plt.figure(figsize=(10,10))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:10])
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```

Highest occuring genres Drama and Comedy followed by action and european.
